```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

# Melakukan Face Detection untuk image object lain

```python
#load cascade
cascade_wajah = cv.CascadeClassifier('/content/drive/MyDrive/Images/haarcascades/haarcascade_frontalface_alt.xml') 
```

```python
#process face image jokowi
jokowi = cv.imread('/content/drive/MyDrive/Images/facedet/jokowi.jpg')
jokowi1 = jokowi.copy()
detec_wajah = cascade_wajah.detectMultiScale(jokowi) 

for(x,y,w,h) in detec_wajah: 
 cv.rectangle(jokowi,(x,y),(x+w,y+h),(0,255,0),3)

#plot gambar
plt.figure(figsize = (10,10)) 
plt.subplot(121),plt.xticks([]),plt.yticks([]),plt.imshow(cv.cvtColor(jokowi1, cv.COLOR_BGR2RGB))
plt.subplot(122),plt.xticks([]),plt.yticks([]),plt.imshow(cv.cvtColor(jokowi, cv.COLOR_BGR2RGB))
plt.show()
```
